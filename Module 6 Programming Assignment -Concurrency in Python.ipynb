13.1 Write the current date as a string to the text file `today.txt`

```python
from datetime import date

today = date.today()
today_str = str(today)

with open('today.txt', 'w') as f:
    f.write(today_str)

print(f'Written to today.txt: {today_str}')
```

Use `date.today()` to get the current date and convert it to a string using `str()`, then write it to `today.txt` using Python's built-in `open()` function with write mode `'w'`.

13.2 Read the text file `today.txt` into the string `today_string`

```python
with open('today.txt', 'r') as f:
    today_string = f.read()

print(f'today_string: {today_string}')
print(f'Type: {type(today_string)}')
```

Open `today.txt` in read mode `'r'` and use `.read()` to load the entire contents into the variable `today_string`.

13.3 Parse the date from `today_string`

```python
from datetime import datetime

parsed_date = datetime.strptime(today_string.strip(), '%Y-%m-%d').date()

print(f'Parsed date: {parsed_date}')
print(f'Type: {type(parsed_date)}')
print(f'Year: {parsed_date.year}, Month: {parsed_date.month}, Day: {parsed_date.day}')
```

Use `datetime.strptime()` to parse the date string back into a `date` object. The format `'%Y-%m-%d'` matches the format written in 13.1 (e.g. `2026-04-24`). We use `.strip()` to remove any whitespace or newline characters.

15.1 Use `multiprocessing` to create three separate processes

```python
import multiprocessing
import time
import random
from datetime import datetime

# Function for each process to run
def print_time(process_num):
    # Wait for a random time between 0 and 1 second
    wait_time = random.random()
    time.sleep(wait_time)
    
    # Get and print the current time
    now = datetime.now().strftime("%H:%M:%S.%f")
    print(f"Process {process_num}: Waited {wait_time:.3f} seconds. Current time: {now}")

if __name__ == "__main__":
    # Create three separate processes
    processes = []
    for i in range(3):
        # We pass (i+1) just to label the processes as 1, 2, and 3
        p = multiprocessing.Process(target=print_time, args=(i + 1,))
        processes.append(p)
        p.start()

    # Wait for all processes to finish
    for p in processes:
        p.join()

    print("All three processes have finished and exited.")
```

Used the `multiprocessing` module to create three processes. Each process uses `random.random()` to wait between 0 and 1 second. After waiting, it prints the current time using `datetime.now()` and then exits. I used `p.join()` to make sure the main program waits for all of them to finish.
